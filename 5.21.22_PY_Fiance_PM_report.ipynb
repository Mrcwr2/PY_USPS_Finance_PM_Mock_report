```python
import pandas as pd
import numpy as np
import csv
import datetime
```

```python
# bring in report from corrigo and remove blank header row
df = pd.read_csv("C:\\Users\\Charles.Robinson\\Desktop\\python reports\\(XX.01.XX - XX.31.XX) Finance - HVAC PMs All status Report-(For Quarter) Mod-04090409.csv", header=1)
df
```

```python
# remove blank columns from corrigo 
df = df.dropna(axis = 1, how = 'all')
df
```

```python
# Show data types
df = df.fillna(0)
df.dtypes
```

```python
# remove $ Signs
total = df.Total.str.replace('$','')
sub = df.Subtotal.str.replace('$','')
nte = df['Service Pro NTE'].str.replace('$', '')
unit = df['Unit Price/Rate'].str.replace('$', '')
print(total)
print(sub)
print(nte)
print(unit)
```

```python
# add the columns with $ removed back to df
df['Total'] = total
df['Subtotal'] = sub
df['Service Pro NTE'] = nte
df['Unit Price/Rate'] = unit
df
```

```python
#  Change columns from str object to float
df['Total'] = pd.to_numeric(df['Total'],errors='coerce')
df['Service Pro NTE'] = pd.to_numeric(df['Service Pro NTE'],errors='coerce')
df['Unit Price/Rate'] = pd.to_numeric(df['Unit Price/Rate'],errors='coerce')
df['Subtotal'] = pd.to_numeric(df['Subtotal'],errors='coerce')
print(df.tail())
```

```python
# find grand totals
total = df.sum().astype(int)
print(total)
```

```python
# Change totals to Dict
total = total.to_dict()
print(total)
```

```python
# append Dict total to dataframe
df = df.append(total, ignore_index=True, sort=False)
df
```

```python
# get todays date and store in vairable 
from datetime import date

today = date.today()
today2 = f'Finance - HVAC PMs All status Report-{today}.xlsx'
```

```python
# Save file
df.to_csv(f'C:\\Users\\Charles.Robinson\\Desktop\\python reports\\{today2}', index = False)
```
